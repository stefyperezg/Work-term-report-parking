```python
census_dir = "C:\\Users\\stefy\\OneDrive - University of Waterloo\Work Term Report 3 - Parking\\Datasets\\Toronto\\Census Tracts\\"
census_file = 'population and dwelling units per DA _ TTS zone_area.csv'
dwell_pop = pd.read_csv(census_dir+census_file)
```

```python
dwell_pop.head(3)
```

```python
dwell_pop['DA Area Fraction'] = dwell_pop['DA Area']/dwell_pop['DA Area_2']
```

```python
dwell_pop = dwell_pop.rename(columns={"Census_dis":'DA_population','Census_d_1':'DA_Total_priv_dwell','Census_pri':'Private_Dwellings_usual_residents','Census_d_2':'DA Land Area','Total priv':'Total_priv_dwell_int'})
dwell_pop.columns
```

```python
# dwell_pop = dwell_pop.drop(['DAUID', 'CTUID', 'CTNAME', 'DA Land Area', 'Census_lat', 'Census_lon','Cut TTS ar', 'ID', 'DAUID_2'], axis=1)
dwell_pop.head(50)
```

```python
dwell_pop_fractioned = dwell_pop
dwell_pop_fractioned['DA_population'] = dwell_pop['DA_population']*dwell_pop['DA Area Fraction']
dwell_pop_fractioned['DA_Total_priv_dwell'] = dwell_pop['DA_Total_priv_dwell']*dwell_pop['DA Area Fraction']
dwell_pop_fractioned['Private_Dwellings_usual_residents'] = dwell_pop['Private_Dwellings_usual_residents']*dwell_pop['DA Area Fraction']
dwell_pop_fractioned = dwell_pop_fractioned.drop('Total_priv_dwell_int', axis=1)
dwell_pop_fractioned.head(60)
```

```python
dwell_pop_fractioned_per_zone = dwell_pop.groupby('gta06',as_index=False).sum()
dwell_pop_fractioned_per_zone.head()
```
